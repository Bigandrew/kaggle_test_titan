```python
#åå­—é‡Œé¢æœ‰æ–‡ç« ï¼Ÿhttp://www.cnblogs.com/north-north/p/4358084.html

import re
train['Title'] = train['Name'].map(lambda x: re.compile(",(.*?)\.").findall(x)[0])
train['Title'] = train['Title'].map(str.strip) 
train['Title'][train.Title=='Jonkheer'] = 'Master'
train['Title'][train.Title.isin(['Ms','Mlle'])] = 'Miss'
train['Title'][train.Title.isin(['Mme','Dona', 'Lady', 'the Countess'])] = 'Mrs'
train['Title'][train.Title.isin(['Capt', 'Don', 'Major', 'Col', 'Sir'])] = 'Mr'
train['Title'][train.Title.isin(['Dr','Rev'])] = 'DrAndRev'
train['Title'] = pd.factorize(train.Title)[0]
train


```

```python
#åŒç†test
test['Title'] = test['Name'].map(lambda x: re.compile(",(.*?)\.").findall(x)[0])
test['Title'] = test['Title'].map(str.strip) 
test['Title'][test.Title=='Jonkheer'] = 'Master'
test['Title'][test.Title.isin(['Ms','Mlle'])] = 'Miss'
test['Title'][test.Title.isin(['Mme','Dona', 'Lady', 'the Countess'])] = 'Mrs'
test['Title'][test.Title.isin(['Capt', 'Don', 'Major', 'Col', 'Sir'])] = 'Mr'
test['Title'][test.Title.isin(['Dr','Rev'])] = 'DrAndRev'
test['Title'] = pd.factorize(test.Title)[0]
test
```

```python
from sklearn.ensemble import RandomForestRegressor
###  å¡«å……ç¼ºå¤±çš„å¹´é¾„èˆ¹èˆ±å±žæ€§


### ä½¿ç”¨ RandomForestClassifier å¡«è¡¥ç¼ºå¤±çš„å¹´é¾„å±žæ€§
def set_missing_ages(df):

    # æŠŠå·²æœ‰çš„æ•°å€¼åž‹ç‰¹å¾å–å‡ºæ¥ä¸¢è¿›Random Forest Regressorä¸­
    age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]

    # ä¹˜å®¢åˆ†æˆå·²çŸ¥å¹´é¾„å’ŒæœªçŸ¥å¹´é¾„ä¸¤éƒ¨åˆ†
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()

    # yå³ç›®æ ‡å¹´é¾„
    y = known_age[:, 0]

    # Xå³ç‰¹å¾å±žæ€§å€¼
    X = known_age[:, 1:]

    # fitåˆ°RandomForestRegressorä¹‹ä¸­
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)

    # ç”¨å¾—åˆ°çš„æ¨¡åž‹è¿›è¡ŒæœªçŸ¥å¹´é¾„ç»“æžœé¢„æµ‹
    predictedAges = rfr.predict(unknown_age[:, 1:])

    # ç”¨å¾—åˆ°çš„é¢„æµ‹ç»“æžœå¡«è¡¥åŽŸç¼ºå¤±æ•°æ®
    df.loc[ (df.Age.isnull()), 'Age' ] = predictedAges 

    return df, rfr

def set_Cabin_type(df):
    df.loc[ (df.Cabin.notnull()), 'Cabin' ] = "Yes"
    df.loc[ (df.Cabin.isnull()), 'Cabin' ] = "No"
    return df

train, rfr = set_missing_ages(train)
train = set_Cabin_type(train)
train
train.info()
```

```python
dummies_Cabin = pd.get_dummies(train['Cabin'], prefix= 'Cabin')

dummies_Embarked = pd.get_dummies(train['Embarked'], prefix= 'Embarked')

dummies_Sex = pd.get_dummies(train['Sex'], prefix= 'Sex')

dummies_Pclass = pd.get_dummies(train['Pclass'], prefix= 'Pclass')

df = pd.concat([train, dummies_Cabin, dummies_Embarked, dummies_Sex, dummies_Pclass], axis=1)
df.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
df
```

```python
test.loc[ (test.Fare.isnull()), 'Fare' ] = 10

test_df = test[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
null_age = test_df[test.Age.isnull()].as_matrix()

X = null_age[:, 1:]
predictedAges = rfr.predict(X)
test.loc[ (test.Age.isnull()), 'Age' ] = predictedAges

test = set_Cabin_type(test)
dummies_Cabin = pd.get_dummies(test['Cabin'], prefix= 'Cabin')
dummies_Embarked = pd.get_dummies(test['Embarked'], prefix= 'Embarked')
dummies_Sex = pd.get_dummies(test['Sex'], prefix= 'Sex')
dummies_Pclass = pd.get_dummies(test['Pclass'], prefix= 'Pclass')


df_test = pd.concat([test, dummies_Cabin, dummies_Embarked, dummies_Sex, dummies_Pclass], axis=1)
df_test.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
df_test
```
